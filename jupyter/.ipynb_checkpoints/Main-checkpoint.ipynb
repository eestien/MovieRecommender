```python
plt.figure(figsize=(20,7))

# Create pd.Series with list of genres for each movie
generlist = movies['genres'].apply(lambda generlist_movie : str(generlist_movie).split("|"))

# Calculate gener frequency value over all dataset
geners_count = {}
for generlist_movie in generlist:
    for gener in generlist_movie:
        if(geners_count.get(gener,False)):
            geners_count[gener]=geners_count[gener]+1
        else:
            geners_count[gener] = 1 

geners_count.pop("(no genres listed)")

geners_count = {k: v for k, v in sorted(geners_count.items(), key=lambda item: item[1], reverse=True)}
# Plot geners frequency values over all data
plt.bar(geners_count.keys(),geners_count.values(),color='b')
```

### Collaborative Filtering: item-based filtering

```python
# One to one maping between movies, user and their rating
ratings_matrix_items = merge_ratings_movies.pivot_table(index=['movieId'],columns=['userId'],values='rating').reset_index(drop=True)
ratings_matrix_items.fillna( 0, inplace = True )
ratings_matrix_items.shape
```

```python
ratings_matrix_items.head()
```

```python
movie_similarity = 1 - pairwise_distances(ratings_matrix_items.to_numpy(), metric="cosine")
np.fill_diagonal(movie_similarity, 0) #Filling diagonals with 0s for future use when sorting is done
ratings_matrix_item_item = pd.DataFrame(movie_similarity)
ratings_matrix_item_item
```

```python
# Below function will take the movie name as a input and will find the movies which are similar to this movie. 
# This function first find the index of movie in movies frame and then take the similarity of movie and align 
# in movies dataframe so that we can get the similarity of the movie with all other movies.


def item_similarity(movieName, ratings_matrix): 
    """
        recomendates similar movies
        :param movieName: name of the movie 
    """
    try:
        user_inp = movieName
        inp = movies[movies['title']==user_inp].index.tolist()
        inp = inp[0]
        
        # Create a copy of movies dataframe
        df_movies = movies
        
        # set new column to movies dataframe: similarity 
        df_movies['similarity'] = ratings_matrix.iloc[inp]
        df_movies.columns = ['movie_id', 'title', 'release_date','similarity']
        return df_movies
    except:
        print("Sorry, the movie is not in the database!")
```

```python
def recommendedMoviesAsperItemSimilarity(user_id, df_movies_ratings, ratings_matrix):
    """
         Recommending movie which user hasn't watched as per Item Similarity
         :param user_id: user_id to whom movie needs to be recommended
         :return: movieIds to user 
    """
    # Get movies for a specific user with a user's highest rating (5 or 4.5)
    user_movie = df_movies_ratings[(df_movies_ratings.userId==user_id) & df_movies_ratings.rating.isin([5,4.5])][['title']]
    
    # Take movie with a highest rating
    user_movie = user_movie.iloc[0,0]
    
    # Get most similar movies based on ratings cosine distance and considering movie name
    df_movies = item_similarity(user_movie, ratings_matrix=ratings_matrix)
    
    # Sort movies based on cosine similarity
    sorted_movies_as_per_userChoice = df_movies.sort_values(["similarity"], ascending = False)
    
    # Set threshold for movie similarity
    sorted_movies_as_per_userChoice = sorted_movies_as_per_userChoice[sorted_movies_as_per_userChoice['similarity'] >=0.45]['movie_id']
    
    # Init df to collect recommended movies
    df_recommended_item=pd.DataFrame()
    
    
    # Define movies that the user has already watched
    user2Movies= ratings[ratings['userId']== user_id]['movieId']
    
    # Check if user haven't seen movie yet
    for movieId in sorted_movies_as_per_userChoice:
            if movieId not in user2Movies:
                df_new= ratings[(ratings.movieId==movieId)]
                df_recommended_item=pd.concat([df_recommended_item,df_new])
    
    # Sort df by rating and get unique movieIds 
    best10 = df_recommended_item.sort_values(["rating"], ascending = False)['movieId'].unique().tolist()[:10]
    return best10
```

```python
def movieIdToTitle(listMovieIDs):
    """
     Converting movieId to titles
    :param user_id: List of movies
    :return: movie titles
    """
    movie_titles= list()
    for id in listMovieIDs:
        
        movie_titles.append(movies[movies['movie_id']==id]['title'].iloc[0])
    return movie_titles
```

```python
user_id=50
print("Recommended movies :\n", movieIdToTitle(recommendedMoviesAsperItemSimilarity(user_id=user_id, df_movies_ratings=merge_ratings_movies, ratings_matrix=ratings_matrix_item_item)))
```

```python
movieId_recommended=list()
def recommendMoviesAsperUserSimilarity(userId):
    """
         Recommending movies which user hasn't watched as per User Similarity
         :param user_id: user_id to whom movie needs to be recommended
         :return: movieIds to user 
    """
    frames  = []
    user2Movies = ratings[ratings['userId']== userId]['movieId']
    sim_user = df_similar_user.iloc[0,0]
    df_recommended=pd.DataFrame(columns=['movieId','title','genres','userId','rating'])
        
    for movieId in ratings[ratings['userId']== sim_user]['movieId']:
        if movieId not in user2Movies:
            df_new= merge_ratings_movies[(merge_ratings_movies.userId==sim_user) & (merge_ratings_movies.movieId==movieId)]
            df_recommended=pd.concat([df_recommended,df_new])
          
    return df_recommended.sort_values(['rating'], ascending = False )['movieId'][:10]
```
